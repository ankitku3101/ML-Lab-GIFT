Imports

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10,100)
y1=np.sin(x)
y2=np.cos(x)
```

Scatter plot

```python
plt.figure(figsize=(8,4))
plt.scatter(x,y1,label='sin(x)',color='red')
plt.scatter(x,y2,label='cos(x)',color='blue')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Scatter plot')
plt.grid(True,'both','both')
plt.legend()
plt.show()
```

Bar Plot

```python
categories = ['A','B','C','D']
values = [10,25,15,30]
plt.figure(figsize=(8,4))
plt.bar(categories,values)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot')
plt.show()
```

Histogram

```python
data = np.random.randn(1000)
plt.figure(figsize=(8,4))
plt.hist(data,bins=30)
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.title('Histogram')
plt.show()
```

Pie Chart

```python
labels=['A','B','C']
sizes=[30,45,25]
plt.figure(figsize=(6,6))
plt.pie(sizes,labels=labels,autopct="%1.1f%%",startangle=90)
plt.title("Pie chart")
plt.show()
```
